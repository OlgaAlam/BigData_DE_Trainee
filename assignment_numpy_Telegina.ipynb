**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
# Generate a random array
arr = np.random.randint(1,15, size=12)
arr
```

```python
# Option 1

# Filter data and multiply by -1
arr[(arr >= 3) & (arr <= 8)] *= -1
arr
```

```python
# Option 2

# Define a mask (negate only where mask is True)
mask = np.where((arr >= 3) & (arr <= 8), True, False)

# Compute negative values where mask is True
res = np.negative(arr, where=mask, out=arr)
res
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
# Generate random arrays
massA = np.random.randint(0, 10, (8,3))
massB = np.random.randint(0, 10, (2,2))
print(massA)
print(massB)
```

```python
def strB_in_strA (massA, massB):
    for RowB in massB:
        for RowA in massA:
            if all(item in RowA for item in RowB):
                print(f'{RowB} in {RowA}')

strB_in_strA(massA, massB)
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
# Generate random arrays
mass = np.random.randint(0, 4, (10,3))

# Compares if adjacent columns have the same values, returning True only if all elements in a row are equal
mass_compared = np.logical_and.reduce(mass[:,1:] == mass[:,:-1], axis=1)

# Creates a new array where not all elements in the row are identical
mass_new = mass[~mass_compared]

print(mass)
print(mass_new)
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
# Generate random arrays
mass = np.random.randint(0, 4, (8, 2))

# Convert each row in array into a tuple
mass_tuple = [tuple(row) for row in mass]

# Remove duplicate rows from the array
mass_res = np.unique(mass_tuple,axis=0)

print(mass)
print(mass_res)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
#Generate random arrays
x = np.random.randint(0, 6, 8)
x
```

```python
# Option 1

def count_repeats_with_lists(x):
    unique_numbers = []
    counts = []

    for num in x:
        if num in unique_numbers:
            index = unique_numbers.index(num)
            counts[index] += 1
        else:
            unique_numbers.append(num)
            counts.append(1)
            
    return (np.array(unique_numbers), np.array(counts))

print(count_repeats_with_lists(x))
```

```python
# Option 2 (numpy)

unic_num = tuple(np.unique(x, return_counts = True))
unic_num
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
# Generate random arrays
X = np.random.randint(1, 6, size=(3,4))
Y = np.random.randint(0, 9, size=(3,4))
print(X)
print(Y)
```

```python
# Option 1 (numpy)

# Calculate the Euclidean distance matrix
dist_matrix = np.sqrt(((X[:, np.newaxis] - Y[np.newaxis, :]) ** 2).sum(axis=2))

print(dist_matrix)
```

```python
#Option 2 (scipy)

from scipy.spatial.distance import cdist
#Calculate the Euclidean distance matrix
dist_matrix = cdist(X, Y, metric='euclidean')

print(dist_matrix)
```
