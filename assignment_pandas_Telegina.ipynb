* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответствующей группы для каждого человека в новой колонке AgeGroup**

```python
data2 = data.copy()
```

```python
cut_labels_3 = ['young', 'adult', 'retiree']
cut_bins = [16, 35, 70, 100]

data2['AgeGroup'] = pd.cut(data2['age'], bins=cut_bins, labels=cut_labels_3)

data2.head()
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
#Data grouped by age group and salary
agegroup_salary = data2.groupby(by=['AgeGroup','salary'], observed=True, as_index=False)[['salary']].value_counts()

#Amounts by age group
sum_by_group = agegroup_salary.groupby(by=['AgeGroup'], observed=True)['count'].transform('sum')

#Percentage of people with >50K and <=50K salary by age group
agegroup_salary['percent'] = (agegroup_salary['count'] / sum_by_group * 100).round(1)

rich = agegroup_salary[agegroup_salary['salary'] == '>50K'].reset_index(drop=True)
rich.drop('percent', axis=1)
```

```python
#Age group including maximum percentage of people with >50K salary
rich[rich['percent'] == rich['percent'].max()].drop(['count', 'salary'], axis=1)
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функцию фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
#Number of people by occupation groups
grouped_occup = data.groupby(by=['occupation'], as_index=False)[['occupation']].value_counts()
grouped_occup
```

```python
# Option 1

#Function for groups filtering
def filter_func1(data):
    #Average age and minimum hours-per-week by occupation groups
    gr = data.groupby(by=['occupation'], as_index=False).agg({'age': 'mean', 'hours-per-week': 'min'}).round(0)
   
    #Occupation set for age filter 
    age_filtered_set = set(gr['occupation'][gr['age'] <= 40])
    
    #Filter occupation groups for hours-per-week conditions 
    filtered_occup_hpw = data.groupby('occupation').filter(lambda x: x['hours-per-week'].min() > 5)
    #Occupation set for hours-per-week filter
    hpw_filtered_set = set(filtered_occup_hpw['occupation'])

    #Set occupation by filtered age and hours-per week
    intersection_set = age_filtered_set.intersection(hpw_filtered_set)
    
    #Excluding data by set occupation
    filtered_data = gr[gr['occupation'].isin(intersection_set)]
    
    return filtered_data
```

```python
#Occupation groups where average age <= 40 and hours-per-week for each person in group > 5
filtered_groups = filter_func1(data)
filtered_groups
```

```python
# Option 2

#Function for groups filtering
def filter_func2(data):
    #Average age and minimum hours-per-week by occupation groups
    gr = data.groupby(by=['occupation'], as_index=False).agg({'age': 'mean', 'hours-per-week': 'min'}).round(0)
   
    #Occupation set for age and hours-per-week filters
    filtered_set = set(gr['occupation'][(gr['age'] <= 40) & (gr['hours-per-week'] > 5)])
    
    #Excluding data by occupation set
    filtered_data = gr[gr['occupation'].isin(filtered_set)]
    
    return filtered_data
```

```python
#Occupation groups where average age <= 40 and hours-per-week for each person in group > 5
filtered_groups = filter_func2(data)
filtered_groups
```
