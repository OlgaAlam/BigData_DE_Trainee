**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
import numpy as np
```

```python
data_marital_salary = data.loc[:,['marital-status', 'salary']]
married_list = 'Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse'
data_marital_salary['status_binary'] = np.where(data_marital_salary['marital-status'].isin(married_list), 'married', 'single')
data_marital_salary.head()
```

```python
data_marital_salary['status_binary'].value_counts()
```

```python
grouped_marital_salary = data_marital_salary.groupby(by=['status_binary', 'salary'], as_index=False)[['salary']].value_counts()
grouped_marital_salary
```

```python
married_percentage = grouped_marital_salary['count'][1] / data_marital_salary['status_binary'].value_counts()[1]
single_percentage = grouped_marital_salary['count'][3] / data_marital_salary['status_binary'].value_counts()[0]
print(f'Доля зарабатывающих доход >50K среди женатых - {married_percentage.round(2)}; среди холостых - {single_percentage.round(2)}.')
```

Ответ: доля зарабатывающих >50K среди женатых значительно больше.

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
data.age.describe()
```

```python
data2 = data.copy()
```

```python
# your code
cut_labels_3 = ['young', 'adult', 'retiree']
cut_bins = [16, 35, 70, 100]

data2['AgeGroup'] = pd.cut(data2['age'], bins=cut_bins, labels=cut_labels_3)

data2.head()
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
# your code
agegroup_salary = data2.groupby(by=['AgeGroup','salary'], as_index=False)[['salary']].value_counts()
agegroup_salary
```

```python
#Группировка по 'AgeGroup'
#Расчет суммы для каждой группы (т.к. нам нужна доля внутри каждой группы)
sum_by_group = agegroup_salary.groupby('AgeGroup')['count'].transform('sum')

#Расчет доли
agegroup_salary['share'] = agegroup_salary['count'] / sum_by_group
agegroup_salary
```

```python
#Фильтрация датасета по доходу >50K
agegroup_50Kmore = agegroup_salary[agegroup_salary['salary'] == '>50K']

#Поиск максимальной доли (признак share) в отфильтрованном датасете 
max_share = agegroup_50Kmore['share'].max()

#Определение значения в признаке AgeGroup с максимальной долей
a = agegroup_50Kmore.loc[agegroup_50Kmore['share'] == max_share, 'AgeGroup']

print(f'Возрастная группа, которая чаще зарабатывает >50К: {a}')
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
# your code
grouped_occupation = data.groupby(by=['occupation'], as_index=False)[['occupation']].value_counts()
grouped_occupation
```

```python
# Определение функции для группировки данных и фильтрации групп
def filter_func(x):
    filtered_data = data.groupby('occupation').filter(lambda x: (x['age'].mean() >= 40) & (x['hours-per-week'].mean() > 5))
    return filtered_data

filtered_groups = filter_func(data)
filtered_groups[10:20]
```
